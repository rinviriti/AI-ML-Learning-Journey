# Understanding Overfitting and Underfitting

One of the biggest challenges in Machine Learning is creating a model that performs well on unseen data.

A model can fail in two ways:

## Underfitting

The model is too simple and cannot learn important patterns from the data.

Characteristics:

- Low training accuracy
- Low testing accuracy

## Overfitting

The model memorizes the training data instead of learning general patterns.

Characteristics:

- Very high training accuracy
- Low testing accuracy

## Goal

Build a model that performs well on both training and testing data.

```python
import pandas as pd

results = pd.DataFrame({
    "Model": ["Underfitting", "Good Fit", "Overfitting"],
    "Training Accuracy": [0.60, 0.88, 0.99],
    "Testing Accuracy": [0.58, 0.85, 0.70]
})

results
```

```python
import matplotlib.pyplot as plt

models = ["Underfitting", "Good Fit", "Overfitting"]

train_acc = [0.60, 0.88, 0.99]
test_acc = [0.58, 0.85, 0.70]

plt.figure(figsize=(8,5))

plt.plot(models, train_acc, marker="o", label="Training Accuracy")
plt.plot(models, test_acc, marker="o", label="Testing Accuracy")

plt.title("Training vs Testing Accuracy")
plt.ylabel("Accuracy")
plt.legend()

plt.show()
```

## Interpretation

### Underfitting

Training Accuracy = 60%

Testing Accuracy = 58%

The model has not learned enough from the data.

### Good Fit

Training Accuracy = 88%

Testing Accuracy = 85%

The model generalizes well.

### Overfitting

Training Accuracy = 99%

Testing Accuracy = 70%

The model memorized the training data and performs poorly on new data.
