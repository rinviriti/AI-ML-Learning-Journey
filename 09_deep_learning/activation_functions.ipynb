# Activation Functions

Activation functions help neural networks learn complex patterns.

Common activation functions:

- ReLU
- Sigmoid
- Softmax
- Tanh

Without activation functions, neural networks would behave like simple linear models.

```python
import numpy as np
import tensorflow as tf

values = tf.constant([-2.0, -1.0, 0.0, 1.0, 2.0])

print(tf.nn.relu(values))
print(tf.nn.sigmoid(values))
```
