# Deep Learning Model Evaluation

Building a model is only the first step.

A model may achieve high performance on training data but fail when exposed to new, unseen data.

Therefore, we must evaluate our model using appropriate metrics.

In this notebook, we will learn:

- Accuracy
- Precision
- Recall
- F1 Score
- Confusion Matrix

These metrics are widely used in Machine Learning, Deep Learning, Medical AI, and Computer Vision.

## Why Model Evaluation Matters

Imagine we develop a skin cancer detection model.

If the model predicts:

- Benign lesions correctly
- Malignant lesions incorrectly

the consequences could be serious.

Therefore, accuracy alone is often not enough.

We need multiple evaluation metrics to understand model performance.

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score
)
```

## Example Predictions

Let's assume:

- y_true = actual labels
- y_pred = model predictions

Where:

0 = Benign
1 = Malignant

```python
y_true = [1, 0, 1, 1, 0, 1, 0, 1]

y_pred = [1, 0, 1, 0, 0, 1, 1, 1]

print("Actual Labels:")
print(y_true)

print("\nPredicted Labels:")
print(y_pred)
```

## Accuracy

Accuracy measures the percentage of correct predictions.

Formula:

Accuracy = Correct Predictions / Total Predictions

Accuracy is useful when the classes are balanced.

```python
accuracy = accuracy_score(y_true, y_pred)

print("Accuracy:", round(accuracy, 3))
```

## Precision

Precision answers:

"When the model predicts positive, how often is it correct?"

Precision is extremely important in applications where false positives are costly.

```python
precision = precision_score(y_true, y_pred)

print("Precision:", round(precision, 3))
```

## Recall

Recall answers:

"Out of all actual positive cases, how many did the model find?"

Recall is particularly important in healthcare.

For example:

If a patient actually has cancer, we want our model to detect as many positive cases as possible.

```python
recall = recall_score(y_true, y_pred)

print("Recall:", round(recall, 3))
```

## F1 Score

F1 Score combines Precision and Recall into a single metric.

It is useful when we need a balance between:

- Precision
- Recall

Many research papers report F1 Score because it provides a more complete picture of model performance.

```python
f1 = f1_score(y_true, y_pred)

print("F1 Score:", round(f1, 3))
```

## Confusion Matrix

A confusion matrix helps us understand exactly where the model makes mistakes.

The matrix contains:

- True Positives (TP)
- True Negatives (TN)
- False Positives (FP)
- False Negatives (FN)

This is one of the most commonly used evaluation tools in Machine Learning.

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_true, y_pred)

print(cm)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.heatmap(
    cm,
    annot=True,
    fmt="d",
    cmap="Blues"
)

plt.title("Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")

plt.show()
```

## Interpretation

Suppose the confusion matrix contains:

- True Positives (TP): Correct cancer detections
- True Negatives (TN): Correct benign detections
- False Positives (FP): Benign predicted as cancer
- False Negatives (FN): Cancer predicted as benign

In healthcare applications, False Negatives are often the most dangerous because a disease may go undetected.
