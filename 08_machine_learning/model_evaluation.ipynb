# Model Evaluation

After training a Machine Learning model, we need to measure how well it performs.

Common metrics include:

- Accuracy
- Precision
- Recall
- F1 Score

These metrics help us understand whether a model is reliable.

```python
from sklearn.metrics import accuracy_score

actual = [1,0,1,1,0]
predicted = [1,0,1,0,0]

accuracy = accuracy_score(actual,predicted)

print("Accuracy:", accuracy)
```
