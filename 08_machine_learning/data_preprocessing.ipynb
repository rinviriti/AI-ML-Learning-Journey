# Data Preprocessing

Raw data is often incomplete, inconsistent, or noisy.

Before training a Machine Learning model, we need to:

- Handle missing values
- Remove duplicates
- Normalize data
- Convert categorical data into numerical form

Data preprocessing is one of the most important steps in Machine Learning.

```python
import pandas as pd

data = {
    "Name": ["Riti", "Rinvi", None],
    "CGPA": [3.86, None, 3.50]
}

df = pd.DataFrame(data)

df
```
