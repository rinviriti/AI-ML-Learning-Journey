# Logistic Regression

Logistic Regression is a classification algorithm.

Instead of predicting numbers, it predicts categories.

Examples:

- Spam vs Non-Spam
- Disease vs Healthy
- Pass vs Fail

```python
from sklearn.linear_model import LogisticRegression

X = [[1],[2],[3],[4],[5]]
y = [0,0,0,1,1]

model = LogisticRegression()

model.fit(X,y)

prediction = model.predict([[4]])

print(prediction)
```
