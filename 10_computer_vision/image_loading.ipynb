# Loading and Displaying Images

Before performing image processing or deep learning, we need to load and visualize images correctly.

In this notebook, we will learn:

- How to load images using OpenCV
- How OpenCV stores images
- How to display images using Matplotlib
- Why RGB and BGR conversion is important

These concepts are fundamental for Computer Vision and Medical AI.

```python
import cv2
import matplotlib.pyplot as plt
```

```python
import cv2
from skimage import data
import matplotlib.pyplot as plt

image = data.astronaut()

print("Shape:", image.shape)

plt.imshow(image)
plt.axis("off")
plt.show()
```

## Loading an Image

Replace `"sample.jpg"` with your own image file.

OpenCV loads images in BGR format by default.

```python
image = cv2.imread("cat.jpg")

print(image.shape)
```

## Understanding Image Dimensions

Images are stored as NumPy arrays.

The shape is represented as:

(height, width, channels)

```python
print("Shape:", image.shape)
```

## Displaying an Image with Matplotlib

Matplotlib expects RGB format.

Therefore, we must convert BGR to RGB.

```python
image_rgb = cv2.cvtColor(
    image,
    cv2.COLOR_BGR2RGB
)

plt.imshow(image_rgb)

plt.title("Loaded Image")

plt.axis("off")

plt.show()
```

## Displaying Image Size

Understanding image dimensions is important because deep learning models usually require fixed-size inputs.

```python
height, width, channels = image.shape

print("Height:", height)
print("Width:", width)
print("Channels:", channels)
```

## Why Image Loading Matters

Every Computer Vision pipeline begins with image loading.

Examples:

- Skin Cancer Detection
- Brain MRI Analysis
- Face Recognition
- Object Detection
- Medical Image Segmentation

A correctly loaded image is the foundation of all later processing steps.
