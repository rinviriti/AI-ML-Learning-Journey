# OpenCV Basics

OpenCV is a popular computer vision library used for image processing, object detection, medical imaging, and real-time vision applications.

In this notebook, we will learn:

- What OpenCV is
- How to import OpenCV
- How images are represented as arrays
- Basic image properties

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

print("OpenCV Version:", cv2.__version__)
```

## Creating a Simple Image

Images are stored as arrays of pixel values.  
Here we will create a blank image using NumPy.

```python
image = np.zeros((300, 300, 3), dtype=np.uint8)

plt.imshow(image)
plt.title("Blank Image")
plt.axis("off")
plt.show()
```

## Drawing on an Image

OpenCV allows us to draw shapes such as lines, rectangles, and circles.

```python
image = np.zeros((300, 300, 3), dtype=np.uint8)

cv2.rectangle(image, (50, 50), (250, 250), (0, 255, 0), 3)
cv2.circle(image, (150, 150), 60, (255, 0, 0), 3)

plt.imshow(image)
plt.title("Shapes with OpenCV")
plt.axis("off")
plt.show()
```
