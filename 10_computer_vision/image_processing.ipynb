# Image Processing with OpenCV

Image processing means modifying or preparing images before using them in Computer Vision or Deep Learning.

In this notebook, we will learn:

- Loading a sample image
- Resizing images
- Converting images to grayscale
- Applying blur
- Detecting edges

These steps are commonly used in medical imaging, object detection, and image classification.

```python
import cv2
import matplotlib.pyplot as plt
from skimage import data
```

## Loading a Sample Image

To avoid file upload errors, we will use a built-in sample image from `skimage`.

This makes the notebook easier for beginners to run.

```python
image = data.astronaut()

plt.imshow(image)
plt.title("Original Image")
plt.axis("off")
plt.show()

print("Image Shape:", image.shape)
```

## Resizing the Image

Deep Learning models usually require fixed image sizes.

For example:

- 224 × 224
- 128 × 128
- 64 × 64

Resizing helps prepare images for model input.

```python
resized_image = cv2.resize(image, (224, 224))

plt.imshow(resized_image)
plt.title("Resized Image")
plt.axis("off")
plt.show()

print("Resized Shape:", resized_image.shape)
```

## Converting to Grayscale

A grayscale image has only one channel instead of three color channels.

Grayscale images are useful when color is not important or when we want to reduce complexity.

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

plt.imshow(gray_image, cmap="gray")
plt.title("Grayscale Image")
plt.axis("off")
plt.show()

print("Grayscale Shape:", gray_image.shape)
```

## Applying Blur

Blurring is used to reduce noise in images.

This can help remove small unwanted details before further processing.

```python
blurred_image = cv2.GaussianBlur(image, (7, 7), 0)

plt.imshow(blurred_image)
plt.title("Blurred Image")
plt.axis("off")
plt.show()
```

## Edge Detection

Edge detection helps identify boundaries in an image.

It is useful for detecting object shapes, outlines, and important structures.

```python
edges = cv2.Canny(gray_image, 100, 200)

plt.imshow(edges, cmap="gray")
plt.title("Edge Detection")
plt.axis("off")
plt.show()
```

## Summary

In this notebook I learned:

- How to load a sample image
- How to resize an image
- How to convert RGB images to grayscale
- How to apply Gaussian blur
- How to detect edges using Canny edge detection

These preprocessing steps are important for Computer Vision and Medical AI projects.
