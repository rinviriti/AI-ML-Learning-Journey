# Image Augmentation

Image augmentation is the process of creating new training images by applying transformations to existing images.

Why is it important?

- Increases dataset size
- Reduces overfitting
- Improves model generalization
- Helps deep learning models learn robust features

Image augmentation is widely used in:

- Medical Imaging
- Skin Cancer Detection
- Brain Tumor Analysis
- Object Detection
- Image Classification

```python
import cv2
import matplotlib.pyplot as plt
from skimage import data
```

## Loading a Sample Image

```python
image = data.astronaut()

plt.imshow(image)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

## Horizontal Flip

Flipping creates a mirror image.

This helps the model learn orientation-independent features.

```python
flipped = cv2.flip(image, 1)

plt.imshow(flipped)
plt.title("Horizontal Flip")
plt.axis("off")
plt.show()
```

## Rotation

Rotating images helps the model become robust to different viewing angles.

```python
height, width = image.shape[:2]

center = (width // 2, height // 2)

rotation_matrix = cv2.getRotationMatrix2D(
    center,
    30,
    1.0
)

rotated = cv2.warpAffine(
    image,
    rotation_matrix,
    (width, height)
)

plt.imshow(rotated)
plt.title("Rotated Image")
plt.axis("off")
plt.show()
```

## Brightness Adjustment

Images may be captured under different lighting conditions.

Adjusting brightness helps models learn under varying illumination.

```python
bright = cv2.convertScaleAbs(
    image,
    alpha=1.0,
    beta=50
)

plt.imshow(bright)
plt.title("Brighter Image")
plt.axis("off")
plt.show()
```

## Zooming

Zoom augmentation helps the model learn from different scales.

```python
h, w = image.shape[:2]

cropped = image[
    h//4:3*h//4,
    w//4:3*w//4
]

zoomed = cv2.resize(
    cropped,
    (w, h)
)

plt.imshow(zoomed)
plt.title("Zoomed Image")
plt.axis("off")
plt.show()
```

## Comparing Augmented Images

Let's compare all transformations together.

```python
fig, axes = plt.subplots(
    1,
    4,
    figsize=(16,4)
)

axes[0].imshow(image)
axes[0].set_title("Original")
axes[0].axis("off")

axes[1].imshow(flipped)
axes[1].set_title("Flip")
axes[1].axis("off")

axes[2].imshow(rotated)
axes[2].set_title("Rotation")
axes[2].axis("off")

axes[3].imshow(bright)
axes[3].set_title("Brightness")
axes[3].axis("off")

plt.show()
```
