# Medical Image Preprocessing

Medical images often contain noise, lighting variations, and unnecessary information.

Before training an AI model, we usually preprocess images to improve quality and consistency.

Common preprocessing techniques include:

- Resizing
- Normalization
- Grayscale Conversion
- Noise Reduction
- Contrast Enhancement

These techniques help deep learning models learn more effectively.

## Why Preprocessing is Important

Imagine two skin lesion images:

- One is bright
- One is dark

Even though both may represent the same condition, the model might struggle to recognize them consistently.

Preprocessing helps reduce these differences and improves model performance.

```python
import cv2
import matplotlib.pyplot as plt
from skimage import data
```

## Loading a Sample Medical-Style Image

For demonstration purposes, we will use a sample image.

In real medical AI projects, this would be:

- MRI
- CT Scan
- X-ray
- Dermoscopic Image

```python
image = data.camera()

plt.imshow(image, cmap="gray")
plt.title("Original Image")
plt.axis("off")
plt.show()
```

## Image Resizing

Deep learning models require fixed image dimensions.

Common sizes:

- 128 × 128
- 224 × 224
- 256 × 256

Resizing ensures consistency across all images.

```python
resized = cv2.resize(image, (224, 224))

plt.imshow(resized, cmap="gray")
plt.title("Resized Image")
plt.axis("off")
plt.show()

print("Resized Shape:", resized.shape)
```

## Normalization

Pixel values usually range from:

0 → 255

Normalization scales them to:

0 → 1

This improves neural network training.

```python
normalized = resized / 255.0

print("Minimum Value:", normalized.min())
print("Maximum Value:", normalized.max())
```

## Noise Reduction

Medical images may contain noise caused by imaging devices.

Gaussian Blur helps reduce unwanted noise.

```python
blurred = cv2.GaussianBlur(
    resized,
    (5,5),
    0
)

plt.imshow(blurred, cmap="gray")
plt.title("Blurred Image")
plt.axis("off")
plt.show()
```

## Contrast Enhancement

Contrast enhancement can highlight important structures.

This is especially useful for:

- MRI
- X-ray
- Skin lesion analysis

```python
equalized = cv2.equalizeHist(resized)

plt.imshow(equalized, cmap="gray")
plt.title("Contrast Enhanced Image")
plt.axis("off")
plt.show()
```

## Comparing Preprocessing Results

Let's compare all preprocessing steps together.

```python
fig, axes = plt.subplots(
    1,
    4,
    figsize=(16,4)
)

axes[0].imshow(image, cmap="gray")
axes[0].set_title("Original")
axes[0].axis("off")

axes[1].imshow(resized, cmap="gray")
axes[1].set_title("Resized")
axes[1].axis("off")

axes[2].imshow(blurred, cmap="gray")
axes[2].set_title("Blurred")
axes[2].axis("off")

axes[3].imshow(equalized, cmap="gray")
axes[3].set_title("Enhanced")
axes[3].axis("off")

plt.show()
```
