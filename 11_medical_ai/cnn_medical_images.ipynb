# CNN for Medical Image Classification

Convolutional Neural Networks are widely used in Medical AI for image classification tasks.

Examples include:

- Skin lesion classification
- Brain tumor detection
- Chest X-ray classification
- Retinal disease detection

In this notebook, we will build a simple CNN workflow using a sample image dataset structure.

This notebook focuses on understanding the medical image classification pipeline.

## Medical Image Classification Workflow

A typical workflow includes:

1. Load medical images
2. Resize images
3. Normalize pixel values
4. Build CNN model
5. Train model
6. Evaluate model
7. Interpret results

This workflow is similar to what is used in projects like MobileSkinXAI.

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

print("TensorFlow Version:", tf.__version__)
```

## Simulated Medical Image Data

For beginner learning, we will use random image tensors to simulate medical image input.

In real projects, these would be actual images such as:

- Skin lesion images
- MRI slices
- X-ray images

Each image will have shape:

64 × 64 × 3

```python
import numpy as np

X_train = np.random.rand(100, 64, 64, 3)
y_train = np.random.randint(0, 2, 100)

X_test = np.random.rand(20, 64, 64, 3)
y_test = np.random.randint(0, 2, 20)

print("Training Images:", X_train.shape)
print("Training Labels:", y_train.shape)
print("Testing Images:", X_test.shape)
print("Testing Labels:", y_test.shape)
```

## Class Labels

For this example:

- 0 = Normal / Benign
- 1 = Abnormal / Disease

This type of binary classification is common in Medical AI.

## Building the CNN Model

The CNN contains:

- Convolution layers to extract image features
- Pooling layers to reduce feature size
- Dense layers for classification
- Sigmoid output for binary prediction

```python
model = Sequential([
    Conv2D(32, (3, 3), activation="relu", input_shape=(64, 64, 3)),
    MaxPooling2D(2, 2),

    Conv2D(64, (3, 3), activation="relu"),
    MaxPooling2D(2, 2),

    Flatten(),

    Dense(128, activation="relu"),
    Dropout(0.5),

    Dense(1, activation="sigmoid")
])

model.summary()
```

## Compiling the Model

For binary medical classification, we use:

- Binary Crossentropy as loss function
- Adam optimizer
- Accuracy as evaluation metric

In real medical AI, we should also check precision, recall, F1-score, and confusion matrix.

```python
model.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)
```

## Training the Model

The model learns patterns from the training images.

Since this is simulated data, the accuracy is not meaningful.

The purpose is to understand the CNN workflow.

```python
history = model.fit(
    X_train,
    y_train,
    epochs=3,
    validation_split=0.2
)
```

## Evaluating the Model

Evaluation checks how the model performs on unseen test data.

In real healthcare projects, evaluation must be very careful because false predictions can have serious consequences.

```python
test_loss, test_accuracy = model.evaluate(X_test, y_test)

print("Test Accuracy:", test_accuracy)
print("Test Loss:", test_loss)
```

## Important Note

This notebook uses simulated random data only for learning CNN structure.

A real medical AI project should use real medical datasets and proper clinical validation.

Examples:

- ISIC dataset for skin lesion classification
