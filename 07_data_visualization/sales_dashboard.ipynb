# Sales Dashboard

In this project, we will analyze and visualize sales data.

Businesses use dashboards to monitor performance, identify trends, and make informed decisions.

## Learning Objectives

- Analyze sales data
- Calculate business metrics
- Create dashboard-style visualizations
- Generate business insights

```python
import pandas as pd

sales = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Sales": [12000, 15000, 18000, 17000, 22000, 25000]
}

df = pd.DataFrame(sales)

df
```

## Monthly Sales Trend

Line charts are useful for showing trends over time.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.plot(
    df["Month"],
    df["Sales"],
    marker="o"
)

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.show()
```

## Total Sales

```python
total_sales = df["Sales"].sum()

print("Total Sales:", total_sales)
```

## Average Monthly Sales

```python
average_sales = df["Sales"].mean()

print("Average Monthly Sales:", average_sales)
```

## Monthly Sales Comparison

```python
plt.figure(figsize=(8,5))

plt.bar(
    df["Month"],
    df["Sales"]
)

plt.title("Monthly Sales Comparison")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.show()
```

## Business Insights

Visualization allows businesses to:

- Track growth trends
- Compare performance over time
- Identify peak sales periods
- Support data-driven decision making
