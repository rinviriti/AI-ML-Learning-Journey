# Bar Chart Visualization

This notebook demonstrates how to create bar charts using Matplotlib.

## Topics Covered

- Creating bar charts
- Comparing categorical data
- Adding labels and titles

```python
import matplotlib.pyplot as plt
```

```python
subjects = ["Math", "Physics", "Chemistry", "Biology"]
marks = [85, 78, 90, 88]

plt.bar(subjects, marks)

plt.title("Student Marks by Subject")
plt.xlabel("Subjects")
plt.ylabel("Marks")

plt.show()
```

```python
subjects = ["Math", "Physics", "Chemistry", "Biology"]
marks = [85, 78, 90, 88]

plt.bar(subjects, marks)

plt.title("Academic Performance")
plt.xlabel("Subjects")
plt.ylabel("Marks Obtained")

for i, value in enumerate(marks):
    plt.text(i, value + 1, str(value))

plt.show()
```
