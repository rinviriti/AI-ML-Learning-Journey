# Introduction to Matplotlib

This notebook introduces the Matplotlib library for creating visualizations in Python.

## Topics Covered
- Importing Matplotlib
- Creating a simple plot
- Adding labels and titles

```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]

plt.plot(x, y)

plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]

plt.plot(x, y)

plt.title("Simple Line Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")

plt.show()
```
