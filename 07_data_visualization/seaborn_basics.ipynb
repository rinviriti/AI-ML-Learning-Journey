# Introduction to Seaborn

Seaborn is a powerful data visualization library built on top of Matplotlib.

While Matplotlib provides basic plotting functionality, Seaborn makes it easier to create attractive and informative statistical visualizations.

Benefits of Seaborn:

- Better visual appearance
- Simpler syntax
- Statistical plotting functions
- Better integration with Pandas

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## Creating a Sample Dataset

```python
tips = sns.load_dataset("tips")

tips.head()
```

## Creating a Scatter Plot with Seaborn

Seaborn can directly use DataFrames, making visualization easier.

```python
sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip"
)

plt.title("Total Bill vs Tip")

plt.show()
```

## Creating a Histogram with Seaborn

```python
sns.histplot(
    data=tips,
    x="total_bill"
)

plt.title("Distribution of Total Bills")

plt.show()
```

## Why Learn Seaborn?

Seaborn is commonly used for:

- Exploratory Data Analysis (EDA)
- Machine Learning Projects
- Research Visualization
- Statistical Analysis

Many Kaggle notebooks and research projects use Seaborn because it provides professional-looking visualizations with very little code.
