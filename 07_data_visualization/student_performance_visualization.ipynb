# Student Performance Visualization

In this project, we will analyze and visualize student performance data using different chart types.

Data visualization helps us understand patterns, trends, and insights that may not be obvious from raw numbers.

## Learning Objectives

- Create visualizations using Matplotlib and Seaborn
- Compare student scores
- Analyze subject performance
- Understand distributions and relationships

```python
import pandas as pd

students = {
    "Student": ["Riti", "Rinvi", "Happy", "Mondira", "Roha"],
    "Math": [85, 90, 78, 72, 88],
    "Physics": [80, 92, 75, 70, 85],
    "Chemistry": [88, 85, 82, 68, 90]
}

df = pd.DataFrame(students)

df
```

## Visualizing Math Scores

Bar charts are useful for comparing values across categories.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.bar(df["Student"], df["Math"])

plt.title("Math Scores")
plt.xlabel("Students")
plt.ylabel("Marks")

plt.show()
```

## Comparing Subject Averages

Let's calculate average scores for each subject.

```python
subject_avg = {
    "Math": df["Math"].mean(),
    "Physics": df["Physics"].mean(),
    "Chemistry": df["Chemistry"].mean()
}

subject_avg
```

```python
plt.figure(figsize=(8,5))

plt.bar(
    subject_avg.keys(),
    subject_avg.values()
)

plt.title("Average Score by Subject")
plt.ylabel("Average Marks")

plt.show()
```

## Score Distribution

Histograms help us understand how scores are distributed.

```python
plt.figure(figsize=(8,5))

plt.hist(df["Math"])

plt.title("Distribution of Math Scores")
plt.xlabel("Marks")
plt.ylabel("Frequency")

plt.show()
```

## Relationship Between Math and Physics Scores

Scatter plots help us identify relationships between variables.

```python
plt.figure(figsize=(8,5))

plt.scatter(
    df["Math"],
    df["Physics"]
)

plt.xlabel("Math")
plt.ylabel("Physics")
plt.title("Math vs Physics Scores")

plt.show()
```
