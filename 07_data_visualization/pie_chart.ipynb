# Pie Chart Visualization

This notebook demonstrates how to create pie charts using Matplotlib.

## Topics Covered

- Creating pie charts
- Displaying proportions
- Adding percentages

```python
import matplotlib.pyplot as plt
```

```python
activities = ["Study", "Sleep", "Exercise", "Entertainment"]
hours = [8, 8, 2, 6]

plt.pie(hours, labels=activities)

plt.title("Daily Activities")

plt.show()
```

```python
activities = ["Study", "Sleep", "Exercise", "Entertainment"]
hours = [8, 8, 2, 6]

plt.pie(
    hours,
    labels=activities,
    autopct="%1.1f%%"
)

plt.title("Daily Time Distribution")

plt.show()
```
