# Histogram Visualization

Histograms are used to understand how numerical data is distributed.

Unlike bar charts, which compare categories, histograms help us analyze the frequency of values within specific ranges (called bins).

For example:

- Student marks
- Age distribution
- Salary ranges
- Medical measurements

Histograms are widely used in Data Science because they help us understand patterns in data before training Machine Learning models.

```python
import matplotlib.pyplot as plt
```

## Sample Dataset

Let's create a simple dataset representing examination scores of students.

```python
marks = [55, 60, 62, 65, 68, 70, 72, 75, 78, 80,
         82, 85, 88, 90, 92, 95]
```

## Creating a Histogram

The histogram groups similar values into bins and shows how many observations fall into each range.

```python
plt.hist(marks)

plt.title("Distribution of Student Marks")
plt.xlabel("Marks")
plt.ylabel("Frequency")

plt.show()
```

## Understanding the Result

From the histogram we can observe:

- Which score ranges occur most frequently
- Whether the data is evenly distributed
- Whether the data contains unusual values

This type of analysis is often performed before applying Machine Learning algorithms.
