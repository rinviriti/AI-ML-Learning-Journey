# Pandas Basics

This notebook introduces the Pandas library for data manipulation and analysis.

## Topics Covered
- Creating Series
- Creating DataFrames
- Viewing data
- Basic DataFrame information

```python
import pandas as pd
```

## Creating a Pandas Series

```python
students = pd.Series(
    ["Rinvi", "Jaman", "Riti"]
)

students
```

## Creating a DataFrame

```python
data = {
    "Name": ["Riti", "Rinvi", "Happy"],
    "Department": ["CSE", "CSE", "EEE"],
    "CGPA": [3.86, 3.85, 3.50]
}

df = pd.DataFrame(data)

df
```

## Viewing the First Rows

```python
df.head()
```

## Viewing Data Information

```python
df.info()
```

## Viewing Summary Statistics

```python
df.describe()
```

## Accessing a Single Column

```python
df["Name"]
```

## Accessing Multiple Columns

```python
df[["Name", "CGPA"]]
```
