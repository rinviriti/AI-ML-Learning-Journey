# Working with DataFrames

This notebook demonstrates common DataFrame operations using Pandas.

## Topics Covered
- Creating DataFrames
- Selecting columns
- Filtering rows
- Sorting data
- Adding new columns
- Saving data as CSV

```python
import pandas as pd
```

```python
students = {
    "Name": ["Riti", "Rinvi", "Happy", "Shopnil"],
    "Department": ["CSE", "CSE", "EEE", "BBA"],
    "CGPA": [3.86, 3.85, 3.75, 3.40]
}

df = pd.DataFrame(students)

df
```

```python
df["Name"]
```

```python
df[["Name", "CGPA"]]
```

```python
high_cgpa = df[df["CGPA"] > 3.80]

high_cgpa
```

```python
sorted_df = df.sort_values(by="CGPA", ascending=False)

sorted_df
```

```python
df["Status"] = ["Excellent", "Excellent", "Good", "Average"]

df
```

```python
df.to_csv("students.csv", index=False)

print("CSV file saved successfully.")
```
